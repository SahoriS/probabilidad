```python
import numpy as np
import scipy as sc
from scipy import stats as st
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

Comenzamos con una introducción general a las variables aleatorias discretas

```python
p = 0.1

bern = st.bernoulli(p)
x = np.arange(-1,3)
f = bern.pmf(x)
print(f)

fig = plt.figure(figsize = (20, 5))
plt.title('Distribución Bernoulli')
plt.plot(x,f)
plt.vlines(x, 0, f)
plt.show()
```

```python
p = 0.5
geo = st.geom(p)
x = np.arange(1,10)
f = geo.pmf(x)

print(f)
fig = plt.figure(figsize = (20, 5))
plt.vlines(x, 0, f)
plt.plot(x, f)
plt.title('Función de masa de probabilidad para una geométrica de parámetro p = 0.5')
plt.show()
```

```python
n, p = 30, 0.2
binom = st.binom(n, p)
x = np.arange(0, n) 
f = binom.pmf(x)
print(f)

fig = plt.figure(figsize = (20,5))
plt.title('Distribución Binomial')
plt.vlines(x, 0, f)
plt.plot(x, f)
plt.show()    
```

Distribución hipergéometrica: 
$$f_{X}(x) = \frac{{r \choose x} {R - r  \choose n - x }}{{R \choose n}}$$

```python
r, R, n = 100, 1000, 20
hip = st.hypergeom(R, r, n)
x = np.arange(0, n)
f = hip.pmf(x)
print(f)

fig = plt.figure(figsize = (20,5))
plt.title('Distribución Hipergeométrica')
plt.vlines(x, 0, f)
plt.plot(x,f)
plt.show()
```

  Distribución Posisson:
  $$f_{X}(x) = \frac{\lambda^{x}}{x!} e^{- \lambda} $$

```python
lamda = 25
poisson = st.poisson(lamda)
x = np.arange(0,100)
f= poisson.pmf(x)
print(f)

fig = plt.figure(figsize = (20, 5))
plt.title('Distribución Poisson')
plt.vlines(x, 0, f)
plt.plot(x, f)
plt.show()
    
```
